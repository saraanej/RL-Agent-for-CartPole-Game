# Training RL Agent for CartPole Game

### **1st Approach:** Using Stable-Baselines3  
Stable-Baselines3 already implements the PPO algorithm, which means that behind the scenes, it's:
- Sampling actions using the current policy
- Collecting rewards
- Estimating how good those actions were (advantages)
- Updating the policy with a clipped objective to avoid wild updates

```python
import os
os.environ["KMP_DUPLICATE_LIB_OK"] = "TRUE"
import matplotlib.pyplot as plt
from stable_baselines3.common.results_plotter import load_results, ts2xy

log_dir = "logs/"
timesteps, rewards = ts2xy(load_results(log_dir), "timesteps")

plt.figure(figsize=(10, 5))
plt.plot(timesteps, rewards)
plt.xlabel("Timesteps")
plt.ylabel("Episode Reward")
plt.title("Training Reward Over Time")
plt.grid()
plt.show()
```
